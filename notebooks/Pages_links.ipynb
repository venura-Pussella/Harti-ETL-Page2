```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

pdf_source = 'https://www.harti.gov.lk/index.php/en/market-information/data-food-commodities-bulletin'

def get_all_pdf_links(pdf_source):
    response = requests.get(pdf_source)
    soup = BeautifulSoup(response.content, 'html.parser')
    pdf_links = soup.find_all('a', href=True)
    
    all_pdf_links = []
    for link in pdf_links:
        if '.pdf' in link['href']:
            full_link = urljoin(pdf_source, link['href'])
            all_pdf_links.append(full_link)
    
    return all_pdf_links

# Get all PDF links
pdf_links = get_all_pdf_links(pdf_source)

# Print all elements in the list
for index, link in enumerate(pdf_links):
    print(f"{link}")
```
