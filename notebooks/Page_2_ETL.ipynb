```python
import tabula
from tabulate import tabulate
import pandas as pd

# Function to read PDF and exptract tables into a DataFrame
def extract_tables_from_pdf_to_df(pdf_path):
    
    tables = tabula.read_pdf(pdf_path, pages='2', multiple_tables=True , lattice=True)
    
    # Combine all tables into a single DataFrame if there are multiple tables
    if isinstance(tables, list):
        combined_df = pd.concat(tables, ignore_index=True)
    else:
        combined_df = tables
    
    return combined_df

pdf_path = '../data/daily_26-08-2024.pdf'

food_df = extract_tables_from_pdf_to_df(pdf_path)

# Print the DataFrame in a tabular format for better readability
print(tabulate(food_df, headers='keys', tablefmt='psql'))

# Optionally save to CSV
# food_df.to_csv('food.csv', index=False)

```

```python
# food_df = food_df.drop(1)
food_df.head(60)
```

```python
import re

# Function to clean the values
def clean_values(value):
    # Use regex to keep only digits, decimal points, and dashes
    return re.sub(r'[^0-9.-]', '', str(value))

# Apply the function to all columns except the first one, excluding the first row
for column in food_df.columns[1:]:
    food_df.loc[1:, column] = food_df.loc[1:, column].apply(clean_values)

food_df.head(60)
```

```python
# food_df.to_csv('food.csv', index=False)
```

```python
# Identify columns with "."
columns_with_dot = [col for col in food_df.columns if '.' in col]

# Extract column names before "." and rename columns
new_column_names = {}
for col in columns_with_dot:
    new_col_name = col.split('.')[0]  # Get part before "."
    new_column_names[col] = new_col_name

# Rename columns
food_df.rename(columns=new_column_names, inplace=True)

# Print the updated DataFrame
food_df.head(60)
```

```python
first_cell_under_the_first_column = food_df.iloc[:, 0]
food_df.iloc[0, 0] = 'Location'
first_cell_under_the_first_column
```

```python
import numpy as np

# Transpose the DataFrame and set the first row as column headers
df_transposed = food_df.T
df_transposed.columns = df_transposed.iloc[0]

df_transposed.head()

# Drop the first row (which is now redundant as column headers) and reset index
df_transposed = df_transposed.drop(df_transposed.index[0])
df_transposed = df_transposed.reset_index().rename(columns={'index': 'Date'})

df_transposed.head()

# # # Attempt to convert using 'YYYY-MM-DD' format
# # df_transposed['Date'] = pd.to_datetime(df_transposed['Date'], format='%Y-%m-%d', errors='coerce')

# # # For any remaining NaT values, attempt conversion using 'M/D/YYYY' format
# # df_transposed['Date'] = df_transposed['Date'].fillna(pd.to_datetime(df_transposed['Date'], format='%m/%d/%Y', errors='coerce'))

# # For any remaining NaT values, attempt conversion using 'DD/MM/YYYY' format
# df_transposed['Date'] = df_transposed['Date'].fillna(pd.to_datetime(df_transposed['Date'], format='%d/%m/%Y', errors='coerce'))

# Clean up the 'Date' column: remove trailing decimal and convert to datetime
df_transposed['Date'] = df_transposed['Date'].str.replace(r'\.\d+$', '', regex=True)
df_transposed['Date'] = pd.to_datetime(df_transposed['Date'], format='%d/%m/%Y', errors='coerce')

df_transposed.head()

df_transposed['Location'] = df_transposed['Location'].str.replace(r'[\r\n]+', ' ', regex=True)
df_transposed.columns.name = None

# Replace empty strings with NaN
pd.set_option('future.no_silent_downcasting', True)
df_transposed.replace(r'^\s*$', float("NaN"), regex=True, inplace=True)
df_transposed.replace(r'^-$', np.nan, regex=True, inplace=True)
# df_transposed.to_csv('food.csv', index=False)
   
# # Get the total number of rows in the DataFrame
# total_rows = len(df_transposed)
# print(total_rows)

# # Identify columns with the same count of NaNs as the total number of rows
# columns_to_drop = [col for col in df_transposed.columns if df_transposed[col].isna().sum() == total_rows]
# print(columns_to_drop)

# # Safely identify columns with the same count of NaNs as the total number of rows
# columns_to_drop = []
# for col in df_transposed.columns:
#     nan_count = df_transposed[col].isna().sum()
#     if nan_count == total_rows:
#         columns_to_drop.append(col)

# print(columns_to_drop)

# # Drop those columns
# df_transposed = df_transposed.drop(columns=columns_to_drop)
# df_transposed.head()

# Reshape the DataFrame from wide to long format
df_melted = df_transposed.melt(id_vars=['Date', 'Location'], var_name='Item_Names', value_name='Value')
df_melted.head()

df_melted['Value'] = df_melted['Value'].replace("-", float("NaN"))

def format_six_digit_number(value):
    if isinstance(value, str) and re.fullmatch(r"\d{6}", value):
        return value[:3] + "-" + value[3:]
    return value

# Apply the function to format 6-digit numbers in the "Value" column
df_melted['Value'] = df_melted['Value'].apply(format_six_digit_number)

# drop the row where value is NaN
df_melted = df_melted.dropna(subset=['Value'])
df_melted = df_melted.dropna(subset=['Item_Names'])

# Display the cleaned DataFrame
df_melted.head(10)

# # # # Save the melted DataFrame to a CSV file
# # df_melted.to_csv('food1.csv', index=False)
```

```python
# Update Item_Names column
df_melted.loc[df_melted['Item_Names'].str.contains('- Small', na=False), 'Item_Names'] = 'Pineapple - Small'
df_melted.loc[df_melted['Item_Names'].str.contains('- Medium', na=False), 'Item_Names'] = 'Pineapple - Medium'
df_melted.loc[df_melted['Item_Names'].str.contains('- Karathakol', na=False), 'Item_Names'] = 'Mango - Karthakolomban'

# # Save the updated DataFrame to a new CSV file
# df_melted.to_csv('updated_food1.csv', index=False)
```

```python
# Ensure all values in 'Value' are strings
df_melted['Value'] = df_melted['Value'].astype(str)

# Split the 'Value' column into 'Min_Value' and 'Max_Value' columns
split_values = df_melted['Value'].str.split('-', expand=True)

# Ensure the split operation results in exactly two columns
if split_values.shape[1] == 2:
    df_melted[['Min_Value', 'Max_Value']] = split_values
else:
    # Handle cases where the split does not result in exactly two columns
    # For example, you might want to log an error or handle it differently
    print("Warning: Split did not result in exactly two columns")

# Convert the new columns to integers, handling potential conversion errors
df_melted['Min_Value'] = pd.to_numeric(df_melted['Min_Value'], errors='coerce').fillna(0).astype(int)
df_melted['Max_Value'] = pd.to_numeric(df_melted['Max_Value'], errors='coerce').fillna(0).astype(int)
```
